```python
import os
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

# 指定包含CSV文件的目錄路徑
input_directory_path = r'D:\TTCIP\SSP126\test'

# 指定輸出目錄的路徑
output_directory_path = r'D:\TTCIP\SSP126\test\output'

# 如果輸出目錄不存在，創建它
os.makedirs(output_directory_path, exist_ok=True)

# 得到所有的.csv文件
csv_files = [f for f in os.listdir(input_directory_path) if f.endswith('.csv')]

# 定義處理單個文件的函數
def process_file(file_name):
    # 創建完整的文件路徑
    input_file_path = os.path.join(input_directory_path, file_name)
    
    try:
        # 讀取.csv文件
        df = pd.read_csv(input_file_path, encoding='utf-8')
        
        # 創建新的文件名稱 (.xlsx)
        new_file_name = os.path.splitext(file_name)[0] + '.xlsx'
        output_file_path = os.path.join(output_directory_path, new_file_name)
        
        # 將數據儲存為.xlsx文件
        df.to_excel(output_file_path, index=False)
        
        print(f'Successfully converted {file_name} to {new_file_name}.')

    except Exception as e:
        print(f'Error converting {file_name}: {str(e)}')

# 使用多執行緒的方式並行處理文件
with ThreadPoolExecutor() as executor:
    executor.map(process_file, csv_files)
```

```python
import os
import pandas as pd

# 設定資料夾路徑
folder_path = r"D:\TTCIP\SSP126\2016"

# 獲取資料夾中所有的CSV檔案路徑
file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 初始化一個空的 DataFrame 用於存放最終結果
combined_df = None

# 處理每一個檔案
for file_path in file_paths:
    # 讀取檔案並將其儲存為 DataFrame
    df = pd.read_csv(file_path)
    
    # 刪除日期欄位為空的行
    df = df.dropna(subset=['DATE'])
    
    # 如果這是第一個檔案，則將其指派給 combined_df
    if combined_df is None:
        combined_df = df
    else:
        # 對於後續的檔案，將它們與 combined_df 進行合併
        # 保留原始的欄位名稱
        # 使用 'inner' 合併來只保留在兩個 DataFrame 中都存在的行
        # 如果新的數據不為 NaN，則用新的數據更新現有的數據
        combined_df = combined_df.combine_first(df)

# 顯示合併後的 DataFrame 的前五行
print(combined_df.head())

# 將合併後的 DataFrame 轉存為 Excel 文件
combined_df.to_excel(r"D:\TTCIP\SSP126\2016\2016_data.xlsx", encoding='utf-8')
```

```python
import os
import shutil

# 設定資料夾路徑
folder_path = r"D:\TTCIP\SSP126"

# 獲取資料夾中所有的CSV檔案路徑
file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 遍歷每個檔案，根據年份建立相應的資料夾並將檔案移動到該資料夾中
for file_path in file_paths:
    # 獲取檔案名稱，例如：'AR6_統計降尺度_日資料_臺灣_平均溫_ssp126_ACCESS-CM2_2015.csv'
    file_name = os.path.basename(file_path)
    # 獲取年份，假設年份在檔案名稱的最後4個字元
    year = file_name[-8:-4]
    # 新建資料夾名稱
    new_folder_name = os.path.join(folder_path, year)
    
    # 如果資料夾不存在，則新建資料夾
    if not os.path.exists(new_folder_name):
        os.makedirs(new_folder_name)
    
    # 將檔案移動到新的資料夾中
    shutil.move(file_path, os.path.join(new_folder_name, file_name))
```

```python
import os
import pandas as pd

# 設定資料夾路徑
folder_path = r"D:\TTCIP\SSP126\test"

# 獲取資料夾中所有的CSV檔案路徑
file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 初始化一個空的 DataFrame 用於存放最終結果
combined_df = None

# Initialize a dictionary to hold dataframes for each model
model_dfs = {}

# 處理每一個檔案
for file_path in file_paths:
    # 讀取檔案並將其儲存為 DataFrame
    df = pd.read_csv(file_path)
    
    # 刪除日期欄位為空的行
    df = df.dropna(subset=['DATE'])
    
    # Extract model name from file name
    model_name = os.path.basename(file_path).split('_')[4]
    
    # If this is the first file of this model, add it to the dictionary
    if model_name not in model_dfs:
        model_dfs[model_name] = df
    else:
        # If this is not the first file of this model, concatenate it with the existing dataframe
        model_dfs[model_name] = pd.concat([model_dfs[model_name], df])

# Combine all model dataframes
for model_df in model_dfs.values():
    if combined_df is None:
        combined_df = model_df
    else:
        combined_df = combined_df.combine_first(model_df)

# 顯示合併後的 DataFrame 的前五行
print(combined_df.head())

# 將合併後的 DataFrame 轉存為 Excel 文件
combined_df.to_excel(r"D:\TTCIP\SSP126\testcombined_data.xlsx", encoding='utf-8')

# 設定資料夾路徑
folder_path = r"D:\TTCIP\SSP126\test"

# 獲取資料夾中所有的CSV檔案路徑
file_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

# 初始化一個空的 DataFrame 用於存放最終結果
combined_df = None

# 處理每一個檔案
for file_path in file_paths:
    # 讀取檔案並將其儲存為 DataFrame
    df = pd.read_csv(file_path)
    
    # 刪除日期欄位為空的行
    df = df.dropna(subset=['DATE'])
    
    # 從檔案中的第四個欄位（模型的數據）提取模型名稱
    model_name = df.columns[3]
    
    # 將第四個欄位重新命名為模型名稱
    df = df.rename(columns={df.columns[3]: model_name})
    
    # 檢查 combined_df 是否為空
    if combined_df is None:
        # 如果是第一個檔案，直接將 DataFrame 賦值給 combined_df
        combined_df = df
    else:
        # 將數據合併到 combined_df 中，使用 'DATE', 'LAT', 'LON' 作為合併鍵
        # 使用 'outer' 連接來保留所有數據並用 NaN 填充缺失值
        combined_df = pd.merge(combined_df, df.drop(columns=['LAT', 'LON']), on=['DATE'], how='outer')

# 顯示合併後的 DataFrame 的前五行
print(combined_df.head())

# 將合併後的 DataFrame 轉存為 Excel 文件
combined_df.to_csv(r"D:\TTCIP\SSP126\testcombined_data.csv", encoding='utf-8')
```
